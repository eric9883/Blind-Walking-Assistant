```python
%cd /
from glob import glob

img_list = glob('/content/dataset/block/export/images/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('content/dataset/block/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list)+'\n')

with open('content/dataset/block/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list)+'\n')
```

```python
import yaml
with open('/content/dataset/block/block.yaml', 'r') as f:
 data = yaml.safe_load(f)

print(data)

data['train'] = '/content/dataset/block/train.txt'
data['val'] = '/content/dataset/block/val.txt'

with open('/content/dataset/block/block.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
